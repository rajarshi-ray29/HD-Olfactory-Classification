```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Convert txt file to csv
df = pd.read_csv('DREAMTrainSet.txt', sep='\t')
df
```

```python
df['Dilution'] = df['Dilution'].str.replace(',', '').apply(pd.eval)
df.to_csv('DREAMTrainSet.csv', index=False) 
df
```

```python
sparsed_df=df.drop(columns=['Replicate','Intensity', 'INTENSITY/STRENGTH', 'subject #'], axis=1)
sparsed_df.dropna(inplace=True)
sparsed_df
```

```python
sparsed_df.to_csv('SparsedTrainSet.csv', index=False) 
```

```python
df=sparsed_df
df.reset_index(drop=True, inplace=True)
df
```

```python
sample_df=df
sample_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(sample_df.drop(columns=['Odor','Compound Identifier'], axis=1))
```

```python
scaled_features=scaler.transform(sample_df.drop(columns=['Odor','Compound Identifier'], axis=1))
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features, columns=sample_df.columns[2:])
df_feat
```

```python
# sns.pairplot(sample_df, hue='Odor')
```

```python
df
```

```python
df=df.iloc[:,:2]
df
```

```python
df = pd.concat([df, df_feat], axis=1)
df
```

```python
df_randomized = df.sample(frac=1, random_state=42)
df_randomized
```

```python
df=df_randomized
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop(columns=['Odor','Compound Identifier'])
y = df['Odor']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', conf_matrix)
```
